```python
# Четырехмерная матрица
A1 = np.array([[3., -5, 2, 4],
              [-3, 1, 0, 1],
              [4, -4, 0, -5],
              [2, 0, 4, -4]])
B1 = np.array([31., -2,-3, 2])
```

```python
A2 = np.array([[-3., 4, 4, 2, 3],
              [3, 2, 0, -1, -5],
              [4, -2, -2, -1, -1],
              [4, 1, 1, -1, 1],
              [4, 4, -4, -1, 2]])
B2 = np.array([14., -27,-14, -15, -49])
```

```python
def print_iteration(A,B):
    for i in range(A.shape[0]):
        print(A[i],'|=|',B[i])
```

```python
print_iteration(A1,B1)
```

```python
def to_triangle(A,B):
    for i in range(A.shape[0] - 1):
        denominator = A[i][i]
        if denominator == 0:
            continue
        A[i] = A[i] / denominator
        B[i] = B[i] / denominator

        for k in range(A.shape[0] - 1, i, -1):
            B[k] = B[k] - A[k][i] * B[i]
            A[k] = A[k] - A[k][i] * A[i]
        print_iteration(A,B)
        print('----------------')
    return A, B
```

```python
def find_roots(A,B):
    X = [0. for i in range(B.shape[0])]

    for i in range(B.shape[0] - 1, -1, -1):
        X[i] = ((B[i] - sum(X * A[i])) / A[i][i]).round()
    print('Найденные корни:')
    return X
```

```python
def Gauss_method(A, B):
    triangle_form = to_triangle(A,B)
    return find_roots(triangle_form[0],triangle_form[1])
```

```python
Gauss_method(A1, B1)
```

```python
Gauss_method(A2, B2)
```
